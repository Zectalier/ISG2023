# Computing the time taken to finish a level.

To compute the time taken to finish a level, we can make the difference between the time of the last 'launched' verb and the time of the first 'completed' verb. We will put the time in a new column called 'time_taken', in a row with the verb 'completed'.

```python
# Load Spy data
clean_spy_data = pd.read_csv('clean_spy_data.csv')

# Remove the unamed columns
clean_spy_data = clean_spy_data.loc[:, ~clean_spy_data.columns.str.contains('^Unnamed')]

# Sort data by actor and timestamp in reverse order
clean_spy_data = clean_spy_data.sort_values(by=['actor', 'timestamp'], ascending=False)
```

```python
# Make a variable to store the last level played.
current_level = None

# Loop through the data
for index, row in clean_spy_data.iloc[::-1].iterrows():
    # If the verb is 'launched' and the current level is not None, 
    # then we store the current level in the variable current_level
    if (row['verbs'] == 'launched' and current_level) or row['verbs'] == 'exited':
        current_level = None

# Save the data of f10c3152 
clean_spy_data[clean_spy_data['actor'] == 'f10c3152'].to_csv('f10c3152.csv')
```

```python
end_time = None
last_completed_index = None

# Iterate over data backwards
for index, row in clean_spy_data.iterrows():
    # If we encounter a 'completed' verb
    if row['verbs'] == 'completed':
        # Store the timestamp as the end time and convert it to a datetime object
        end_time = pd.to_datetime(row['timestamp'])
        
        # Store the index of the last completed verb
        last_completed_index = index

    # If we encounter a 'launched' verb
    if row['verbs'] == 'launched':
        # If we have an end time
        if end_time:
            # Store the start time and convert it to a datetime object
            start_time = pd.to_datetime(row['timestamp'])
            # Calculate the time taken and store it in the dataframe with the last completed verb
            clean_spy_data.at[last_completed_index, 'time_taken'] = (end_time - start_time).total_seconds()
            
            # Reset the end time
            end_time = None
```

```python
# Load full spy data
spy_data = pd.read_csv('full_spy_data.csv')

# Temporary list of actors
actors = []

# Iterate over data 
for index, row in spy_data.iterrows():
    # Create a json object from the actor column
    a = json.loads(row['actor'])
    # Append the name of the actor to the list
    actors.append(a['name'])

spy_data['actor'] = actors

spy_data.head()
```

```python
# Temporary list of verbs
verbs = []

# Iterate over data
for index, row in spy_data.iterrows():
    # Create a json object from the verb column
    a = json.loads(row['verb'])
    # Append the name of the verb to the list
    verbs.append(a['display']['en-US'])

spy_data['verb'] = verbs
```

```python
spy_data.head()
```

```python
# Sort data by actor and timestamp in reverse order
spy_data = spy_data.sort_values(by=['actor', 'timestamp'], ascending=False)
```

```python
all_levels = []
current_level = None

# Iterate over the actors
for actor in spy_data['actor'].unique():
    # Temporary list of levels
    levels = []
    # Iterate over the data of the actor backwards
    for index, row in spy_data[spy_data['actor'] == actor].iloc[::-1].iterrows():
        
        # If the verb of clean_spy_data using the index is 'launched'
        if row['verb'] == 'launched':
            # Convert the object column to a json object
            a = json.loads(row['object'])

            # Set the current level to the extension name without '.xml' at the end
            current_level = a['definition']['extensions']['https://spy&46;lip6&46;fr/xapi/extensions/value'][0][:-4]
            
        levels.append(current_level)
    
    all_levels.append(levels[::-1])
```

```python
# Add the levels backwards to the dataframes
spy_data['levels'] = all_levels
```

```python
# Create a new dataframe with the columns we want from spy_data: actor, timestamp, stored, verb, levels
clean_spy_data = pd.DataFrame()
clean_spy_data['actor'] = spy_data['actor']
clean_spy_data['timestamp'] = spy_data['timestamp']
clean_spy_data['stored'] = spy_data['stored']
clean_spy_data['verb'] = spy_data['verb']
clean_spy_data['levels'] = spy_data['levels']
clean_spy_data['result'] = spy_data['result']
clean_spy_data['time'] = spy_data['time']

clean_spy_data.head()
```
